```python
import matplotlib.pyplot as plt

waktu = [0.0483, 0.0735, 0.0877, 0.1214, 0.1436,
         0.2345, 0.4064, 0.6040, 0.8047, 0.9895, 1.8821]

data = ['10K', '20K', '30K', '40K', '50K',
        '100K', '200K', '300K', '400K', '500K', '1M']

plt.figure(figsize=(12,7))

plt.plot(
    data,
    waktu,
    marker='o',
    markersize=8,
    linewidth=3,
    color='green'
)

plt.title(
    "Perbandingan Jumlah Data vs Waktu Eksekusi MPI (n=8)",
    fontsize=18,
    fontweight='bold'
)

plt.xlabel(
    "Jumlah Data",
    fontsize=13
)

plt.ylabel(
    "Waktu Eksekusi (detik)",
    fontsize=13
)

plt.grid(
    True,
    linestyle='--',
    alpha=0.5
)

for i, y in enumerate(waktu):
    plt.text(
        i,
        y + 0.03,
        f"{y:.3f}",
        ha='center',
        fontsize=10
    )

# MEMPERBESAR TICK
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)

# MARGIN
plt.tight_layout()

plt.show()
```
