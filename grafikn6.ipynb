```python
import matplotlib.pyplot as plt

waktu = [0.0409, 0.0625, 0.0862, 0.1181, 0.1486,
         0.2406, 0.4102, 0.6189, 0.8267, 0.9907, 1.9587]

data = ['10K', '20K', '30K', '40K', '50K',
        '100K', '200K', '300K', '400K', '500K', '1M']

plt.figure(figsize=(12,7))

plt.plot(
    data,
    waktu,
    marker='o',
    markersize=8,
    linewidth=3,
    color='red'
)

plt.title(
    "Perbandingan Jumlah Data vs Waktu Eksekusi MPI (n=6)",
    fontsize=18,
    fontweight='bold'
)

plt.xlabel(
    "Jumlah Data",
    fontsize=13
)

plt.ylabel(
    "Waktu Eksekusi (detik)",
    fontsize=13
)

plt.grid(
    True,
    linestyle='--',
    alpha=0.5
)

for i, y in enumerate(waktu):
    plt.text(
        i,
        y + 0.03,
        f"{y:.3f}",
        ha='center',
        fontsize=10
    )

# MEMPERBESAR TICK
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)

# MARGIN
plt.tight_layout()

plt.show()
```
