```python
import matplotlib.pyplot as plt
import numpy as np

jumlah_data = ['100.000', '200.000', '300.000', '400.000', '500.000', '1.000.000']

waktu_1  = np.array([0.4079, 0.7429, 1.0180, 1.4387, 1.3978, 2.6784])
waktu_2  = np.array([0.4244, 0.8360, 1.0098, 1.4045, 1.5812, 2.7800])
waktu_4  = np.array([0.4163, 0.7461, 1.1518, 1.3031, 1.4022, 2.5374])
waktu_6  = np.array([0.4735, 0.7575, 1.0180, 1.3887, 1.4251, 2.4767])
waktu_8  = np.array([0.4346, 0.7842, 1.0838, 1.3230, 1.4425, 2.7599])
waktu_10 = np.array([0.3674, 0.7163, 1.0129, 1.2418, 1.3160, 2.4810])

plt.figure(figsize=(13,7))

# kasih offset Y biar ga numpuk
offset = 0.08

plt.plot(jumlah_data, waktu_1,            marker='o', linewidth=2, label='1 Proses')
plt.plot(jumlah_data, waktu_2 + offset,   marker='s', linewidth=2, label='2 Proses')
plt.plot(jumlah_data, waktu_4 + offset*2, marker='^', linewidth=2, label='4 Proses')
plt.plot(jumlah_data, waktu_6 + offset*3, marker='D', linewidth=2, label='6 Proses')
plt.plot(jumlah_data, waktu_8 + offset*4, marker='x', linewidth=2, label='8 Proses')
plt.plot(jumlah_data, waktu_10+ offset*5, marker='*', linewidth=2, label='10 Proses')

plt.title("Perbandingan Jumlah Data terhadap Waktu Eksekusi TOPSIS",
          fontsize=16, fontweight='bold')

plt.xlabel("Jumlah Data")
plt.ylabel("Waktu Eksekusi (detik)")
plt.grid(True, linestyle='--', alpha=0.4)

plt.legend(title="Jumlah Proses MPI")
plt.ylim(0.3, 3.2)

plt.tight_layout()
plt.show()
```
